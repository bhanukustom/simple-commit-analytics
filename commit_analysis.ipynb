```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Use the below command to generate csv
`
git log 
    --since='2 year ago' 
    --author=bhanu
    --date=format-local:"%Y-%m-%d %H:%M:%S, %w, %m, %H"
    --pretty=format:%ad > my_commits.csv
`

```python
ch = pd.read_csv('my_commits.csv', names=["DATE", "WEEK","MONTH", "HOUR"])
print(ch.head())

# git log always returns latest first
commits = min(ch['DATE'].tail(1)), max(ch['DATE'].head(1))

monthly = ch['MONTH'].value_counts().sort_index()

weekly = ch['WEEK'].value_counts().sort_index()

hourly = ch['HOUR'].value_counts().sort_index()

fig=plt.figure(figsize=(20, 6), dpi=180, facecolor='w', edgecolor='k')

sp1 = plt.subplot(1, 2, 1)
sp1.set_ylabel('COMMITS')
sp1.set_xlabel('MONTHLY')
monthly.plot.bar(color='deeppink', edgecolor='orange')
plt.xticks(rotation='horizontal')

sp2 = plt.subplot(1, 2, 2)
sp2.set_xlabel('WEEKYLY')
weekly.plot.bar(color='orange', edgecolor='deeppink')
plt.xticks(rotation='horizontal')

plt.suptitle('Code commits from %s to %s'% (commits[0], commits[1]), fontsize=26)
```
