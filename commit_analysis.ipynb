```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Use the below command to generate csv
`
git log 
    --since='2 year ago' 
    --author=bhanu
    --date=format-local:"%Y-%m-%d %H:%M:%S, %w, %m, %H"
    --pretty=format:%ad > my_commits.csv
`

```python
ch = pd.read_csv('all_commits.csv', names=["DATE", "WEEK","MONTH", "HOUR", "AUTHOR"])
# print(ch.head())
# git log always returns latest first
commits = min(ch['DATE'].tail(1)), max(ch['DATE'].head(1))

monthly = ch['MONTH'].value_counts().sort_index()
akash_monthly = ch[ch['AUTHOR'] == 'Akash']['MONTH'].value_counts().sort_index()
bhanu_monthly = ch[ch['AUTHOR'] == 'Bhanu']['MONTH'].value_counts().sort_index()

akash_weekly = ch[ch['AUTHOR'] == 'Akash']['WEEK'].value_counts().sort_index()
bhanu_weekly = ch[ch['AUTHOR'] == 'Bhanu']['WEEK'].value_counts().sort_index()
# weekly = ch['WEEK'].value_counts().sort_index()

akash_hourly = ch[ch['AUTHOR'] == 'Akash']['HOUR'].value_counts().sort_index()
bhanu_hourly = ch[ch['AUTHOR'] == 'Bhanu']['HOUR'].value_counts().sort_index()
# hourly = ch['HOUR'].value_counts().sort_index()


fig=plt.figure(figsize=(20, 6), dpi=180, facecolor='w', edgecolor='k')

sp1 = plt.subplot(1, 2, 1)
sp1.set_ylabel('COMMITS')
sp1.set_xlabel('MONTHLY')
akash_monthly.plot.bar(color='deeppink', edgecolor='orange')
# bhanu_monthly.plot.bar(color='green', edgecolor='orange')
plt.xticks(rotation='horizontal')

sp2 = plt.subplot(1, 2, 2)
sp2.set_xlabel('WEEKYLY')
akash_weekly.plot.bar(color='orange', edgecolor='pink')
plt.xticks(rotation='horizontal')

plt.suptitle('Code commits from %s to %s'% (commits[0], commits[1]), fontsize=26)
```
